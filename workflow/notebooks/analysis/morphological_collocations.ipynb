## Build the Counts

```python
# convert morph nodes into boolean values
morphs = [
    'PL', 'DUAL', 'SFX:1', 
    'SFX:2', 'SFX:3'
]

morph_bool = 1*(~df_sg[morphs].isna())

morph_df = df[['times_utf8', 'TA Heads', 'times_utf8d']].join(morph_bool, how='right')

morph_df['Ø'] = 1*(morph_df[morphs].sum(1) == 0)
```

```python
morph_df
```

```python
# prune to top x occurring terms
lex_ct = morph_df.times_utf8.value_counts()

lex_ct.head(30)
```

```python
top_lexs = lex_ct[lex_ct > 4]
```

```python
morph_ct = pd.pivot_table(
    morph_df[morph_df.times_utf8.isin(top_lexs.index)],
    index=['times_utf8', 'TA Heads', 'times_utf8d'],
    columns=morphs+['Ø'],
    aggfunc='size',
    fill_value=0,
)

print(morph_ct.shape)

morph_ct.droplevel(1, axis=0)
```

```python
advbs = (morph_ct[(0, 0, 0, 0, 0, 1)] - morph_ct.sum(1)) == 0
advbs = advbs[advbs]
pd.DataFrame(advbs).droplevel([1, 2], 0)
```

```python
morph_ct2 = morph_ct.drop(index=advbs.index)

morph_ct.droplevel([1,2], 0)
```

```python
morph_ct.droplevel([1,2], 0)
```

```python
# todo: Replace with automatically determined headers
# headers = pd.DataFrame([
#     ['no modifier', 'third person suffix']
# ]
# )

headers = ['Ø', 'SFX3', 'SFX2', 'SFX1', 'DU', 'PL', 'PL+SFX3', 'PL+SFX2', 'PL+SFX1']
```

```python
# Export the table of raw counts


out.table(
    morph_ct2.droplevel([0,2], 0).set_axis(headers, 1), 
    'morph_counts',
    adjustbox=True,
    caption='Time Adverbial Morpheme Collocations (raw count)',
    label='table:morphcount',
    multicolumn=False
)
```

```python
# Apply deltaP
morph_dp = sig.apply_deltaP(morph_ct2, 0, 1)
morph_dp.droplevel([1,2],0)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
heatmap(morph_dp.droplevel([0,1], 0), ax=ax, square=False, annot=True)
ax.set_yticklabels(ax.get_yticklabels(), size=14, fontfamily='SBL Biblit')
ax.set_ylabel('Time Adverbial Head Lexeme')
ax.set_xticklabels(headers, rotation=20)
ax.set_xlabel('Collocating Morphemes')
out.plot('headXmorph_ΔP', format='pdf')
```
