## Build definitions of modified / non-modified

```python
morphs = ['PL', 'DU'] #'SFX1', 'SFX2', 'SFX3']
df_sg['Ø'] = 1*(df_sg[morphs].sum(1) == 0)
df_sgph['Ø'] =  1*(df_sgph[morphs].sum(1) == 0)
morphs.append('Ø')
```

## Build the Counts

```python
# prune to top x occurring terms
lex_ct = df_sg.times_etcbc.value_counts()

lex_ct.head(30)
```

```python
top_lexs = lex_ct[lex_ct > 4]
```

```python
morphct_ph = pd.pivot_table(
    df_sgph[
        (df_sgph.heads_etcbc.isin(top_lexs.index))
        & (df_sgph.function != 'Time')
    ],
    index=['heads_utf8', 'Phrase Heads', 'heads_utf8d'],
    columns=morphs,
    aggfunc='size',
    fill_value=0,
)

morphct_tm = pd.pivot_table(
    df_sg[df_sg.times_etcbc.isin(top_lexs.index)],
    index=['times_utf8', 'TA Heads', 'times_utf8d'],
    columns=morphs,
    aggfunc='size',
    fill_value=0,
)

headers = ['Ø', 'DU', 'PL']
morphct_tm.columns = headers
morphct_ph.columns = headers

# sort them
morphct_tm = morphct_tm[
    morphct_tm.sum().sort_values(ascending=False).index
]
morphct_tm = morphct_tm.loc[
    morphct_tm.sum(1).sort_values(ascending=False).index
]
morphct_ph = morphct_ph[
    morphct_tm.sum().sort_values(ascending=False).index
]
```
