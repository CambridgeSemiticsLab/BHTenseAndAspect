### Single Phrase Constructions

```python
singles = []
multi = []

for tc in F.label.s('timephrase'):
    
    nphrases = L.d(tc, 'phrase')
    nphraseatoms = L.d(tc, 'phrase_atom')
    time = Time(tc, api)
    times = time.times
    preps = time.preps
    
    if not times:
        raise Exception(tc)

    if all([len(nphrases) == 1, 
            len(nphraseatoms) == 1,
            len(times) == 1,
            len(preps) in {0, 1}]):
        singles.append(tc)
        
    else:
        multi.append(tc)
        
print(f'{len(singles)} single phrasal constructions found...')
print(f'{len(multi)} multi phrasal constructions found...')
```

# Classifications

The classifications are added to a dictionary, keyed by name, to keep track of which constructions are accounted for. The process is one of elimination and deduction based on features identified in an inductive analysis.

```python
classes = {}
```

```python
total_cxs = len(singles) + len(multi)

def percent(n, total):
    return round(n/total, 2)

def prog():
    '''Reports progress in %'''
    found = set(cx for cls, clset in classes.items() for cx in clset)
    ratio = percent(len(found), total_cxs)
    print(f'\t{ratio} ({len(found)}) now accounted for')
    
def get_remaining(cx_list):
    '''Returns remaining adverbials 
    (not classified) from a list'''
    found = set(cx for cls, clset in classes.items() for cx in clset)
    return set(cx_list) - found

def remaining_tags(cx_list):
    '''
    Provides a count of the remaining construction tags 
    that have yet to be accounted for.
    '''
    remaining = get_remaining(cx_list)
    rem_tags = collections.Counter()
    for cx in remaining:
        time = Time(cx, api)
        rem_tags[time.tag] += 1
    return convert2pandas(rem_tags)

def get_cx(node):
    '''Returns a chunk node'''
    return [ch for ch in L.u(node, 'chunk') if F.label.v(ch) == 'timephrase'][0]

class CXdata:
    '''
    Makes count and result data available on
    a given construction class. The class is
    identified through a supplied search function.
    '''
    def __init__(self, cx_set, validate, tokenize):
        
        self.timeset = set()
        counts = collections.Counter()
        counts2d = collections.defaultdict(lambda:collections.Counter())
        self.results = collections.defaultdict(list)
    
        for cx in cx_set:
            time = Time(cx, api)
            if validate(time):
                
                count_text = tokenize(time)
                if type(count_text) == str:
                    counts[count_text] += 1
                elif type(count_text) == tuple: # handle 2 dimensional counts
                    counts[count_text[0]] += 1
                    counts2d[count_text[1]][count_text[0]] += 1
                    
                self.timeset.add(cx)
                result = [cx] + list(L.d(cx, 'word'))
                self.results[count_text].append(result)
                self.results['all'].append(result)
                
        self.counts = convert2pandas(counts)
        self.counts2d = pd.DataFrame(counts2d).fillna(0)
        
        print(f'{len(self.timeset)} matching constructions found...')
        
    def show(self, **options):
        showtag = options.get('tag', 'all')
        shuffle = options.get('shuffle', False)
        showme = list(iter(self.results[showtag]))
        if shuffle:
            random.shuffle(showme)
        options = {k:v for k,v in options.items()
                      if k not in ('showtag', 'shuffle')}
        A.show(self.results[showtag], **options)
```

### The Non-Prepositional (ø) "Adverb Construction"


The adverb construction is a single-word construction that is anchored to speech time. The lack of any specifiers such as demonstratives, definite articles, and other specifications means that these words must necessarily be anchored to an implied common point. This gives rise to a construction that can be considered an "adverb construction." Is this phenomenon discussed in the literature? Do other scholars recognize the association between modifiers and anchoredness?

These constructions are selected simply by finding those with only one specification: `time`.

```python
def check_advb(time):
    '''Checks for adverb status'''
    if time.tag == 'time':        
        return True

def token_advb(time):
    '''Tokenizes adverb cx'''
    text = tokenize_surface(time.cx, api)
    return text

np_adverbs = CXdata(singles, check_advb, token_advb)

classes['ø_adverb'] = np_adverbs.timeset

prog()
```

### The ø Construct Anchor Construction

Times can be anchored to a noun, verb, or suffix that is recoverable from the context. Anchors possibly occur across many main time construction types (as shown below). The patterns identified below consist of a basic `time + anchor` sequence. This pattern is different from the adverb construction identified above in that they are not inherently deictic. That is, the time anchor is specified, not assumed. Yet the pattern is distinct from the Demonstrative Heh construction and the Attributive Anchor construction.

```python
def val_ancacc(time):
    '''Validates an anchored accusative construction'''
    
    this_time = time.times[0]
    p1 = this_time+1 if F.lex.v(this_time+1) != 'H' else this_time+2
    if all([time.tag.startswith('time.construct'),
            F.nu.v(p1) == 'sg' or F.pdp.v(p1) == 'verb']):
        return True
    
def token_ancacc(time):
    return F.lex_utf8.v(time.times[0])

ancacc = CXdata(singles, val_ancacc, token_ancacc)
```

### The Demonstrative Heh Construction

This construction is unique, since the definite article heh is construed as a demonstrative. This is not a typical function of the definite article in noun phrases. It uniquely takes on this role in time constructions.

```python
def val_demheh(time):
    '''Validates a demonstrative heh cx'''
    if time.tag == 'time.H':
        return True
    
def token_demheh(time):
    '''Tokenizes demonstrative heh cx'''
    this_time = time.times[0]
    heh = this_time - 1
    return f'{F.lex_utf8.v(heh)}.{F.lex_utf8.v(this_time)}'

demheh = CXdata(singles, val_demheh, token_demheh)

classes['ø_demon_heh'] = demheh.timeset

prog()
```

### The ø "Attributive Anchor" Construction, [H + time + H + anchor]

This construction takes advantage of the  noun construction. It frequently occurs with the demonstrative, in which case the demonstrative serves to anchor the time word.


```python
def check_atanchor(time):
    '''Validates attributive anchor construction'''
    if 'attr_patt' in time.specs and 'time' in time.specs:
        return True
    
def token_atanchor(time):
    '''Tokenize attributive anchor cx'''
    thetime = time.times[0]
    attrib = thetime + 2
    return F.g_cons_utf8.v(attrib)

attimes = CXdata(singles, check_atanchor, token_atanchor)

classes['ø_attrib_anchor'] = attimes.timeset

prog()
```

### The "Quantified Durative" Construction

```python
def qd_validate(time):
    '''
    Validates a quantified durative cx
    Selects quants but excludes
        • qualitatives
        • the number "one"
    '''

    quant_lexs = set(F.lex.v(q) for q in time.quants) # to check for אחד
    if all(['time' in time.specs, 
            'quant' in time.specs,  
            'card' in time.specs,
            quant_lexs - {'>XD/'}]):    
        return True
    
def token_qd(time):
    '''Tokenizes quantative durative cxs'''
    time_text = F.lex_utf8.v(time.times[0])
    return time_text
    
qdtimes = CXdata(singles, qd_validate, token_qd)
        
classes['ø_quant_durative'] = qdtimes.timeset
        
prog()
```

### The Construct Durative Time Units Construction

Some constructions consist of a time + construct + time units, construing a given time as a distributed and quantified entity. This is most frequently with יום as the time unit.

```python
def val_durtu(time):
    '''Validates a durative time unit construction'''
    this_time = time.times[0]
    p1 = this_time + 1
    
    if all(['time' in time.specs,
            not {'quant'} & set(time.specs),
            'construct' in time.specs,
            F.nu.v(p1) == 'pl',
            F.pdp.v(p1) == 'subs',
            F.prs.v(p1) in {'absent', 'NA'}]):
        return True
    
def token_durtu(time):
    '''Tokenizes a durative time unit cx'''
    return F.lex_utf8.v(time.times[0])

durtu = CXdata(singles, val_durtu, token_durtu)

classes['ø_timeunit_durative'] = durtu.timeset

prog()
```

## Check for Overlap

Which classes overlap and where?

```python
overlaps = collections.defaultdict(list)

for cxclass, timeset in classes.items():
    for cxclass2, timeset2 in classes.items():
        if cxclass == cxclass2:
            continue
        
        union = f'{cxclass}|{cxclass2}'    
        overlapping = timeset & timeset2
        
        if overlapping:
            overlaps[union] = [[cx]+list(L.d(cx, 'word')) for cx in overlapping]
            
overlaps
```

I see currently about 5 prominent types.

* coordinate direction constructions - these are cases of stacked prepositions, where each preposition further specifies the location of the time. There are numerous different species of this CX.
* coordinate adverb constructions - e.g. יומם ולילה
* combined count constructions - This is closely related to the coordinate adverb construction. e.g. שׁתים ושׁמונים שׁנה ושבע מאות שׁנה
* apposition constructions - phrase B further specifies the time quality or location of phrase A; there are also cases of aposition where B modifies not the time but a noun anchor in the time phrase A
* What are cases like יום ביום such as 1 Sam 18:10 doing?

I predict that the most common construction is the coordinate direction CX. We will start with that one so as to begin eliminating possibilities.

### Composite Direction Construction

Cases of stacked prepositions which locate and relocate the position in time

```python
def val_comdir(time):
    '''Validates a coordinate direction CX'''
    
    if len(time.preps) > 1:
        
        # ensure a prepositional pair
        # is not interrupted with waw
        for i, prep in enumerate(time.preps):
            
            # stop on last 
            if i == len(time.preps)-1:
                continue
            
            # look between preps for conjunction
            nxprep = L.d(time.preps[i+1], 'word')
            prepwrds = L.d(prep, 'word')
            iwords = list(range(prepwrds[0], nxprep[0]))
            is_conj = 'conj' in set(F.pdp.v(w) for w in iwords)
            is_quant = 'card' in set(F.ls.v(w) for w in iwords)
            if not is_conj or is_quant:
                return True
            
    
# def tok_cordir(time):
#     '''Tokenize coordinate direction cx in 2D'''
    
cordir = CXdata(multi, val_comdir, token_advb)
```

#### Repeating Durative Construction

The cases above contain some false positives, such as בבקר בבקר in Exod 30:7. These do not represent cases of composite direction, wherein preposition B locates preposition A. Rather, they are cases of repetition to convey duration. In order to isolate these cases from the composite direction cases, I need a function that validates such cases. 

```python
def longest_word_match(time):
    '''
    Credit to stack overflow user Arvind Padmanabhan
    https://codereview.stackexchange.com/questions/63329/finding-the-largest-repeating-substring
    '''
    
    l = [F.g_cons.v(w) for w in L.d(time.cx, 'word')]
    d = collections.deque(l[1:])
    match = []
    longest_match = []
    while d:
        for i, item in enumerate(d):
            if l[i]==item:
                match.append(item)
            else:
                if len(longest_match) < len(match):
                    longest_match = match
                match = []
        d.popleft()
        
    return longest_match

def short_match(time):
    '''
    Matches a phrase with only two words
    that are equivalent.
    '''
    
    words = L.d(time.cx, 'word')
    
    conds = (
        len(words) == 2,
        F.g_cons.v((words[0])) == F.g_cons.v((words[1]))
    )
    
    if all(conds):
        return True
```

```python
def val_repeat(time):
    '''
    Validates a repeating durative construction
    The function looks for identical stretches
    of strings within a time CX.
    '''
    
    # get longest repeated string
    long_match = longest_word_match(time)
    
    conds = (
        len(long_match) == len(L.d(time.cx, 'word'))/2,
        short_match(time)
    )
        
    if any(conds):
        return True
```

Now we re-implement the solution into the composite direction construction algorithm.

```python
def val_comdir(time):
    '''Validates a coordinate direction CX'''
    
    if len(time.preps) > 1:
        
        # ensure a prepositional pair
        # is not interrupted with waw
        for i, prep in enumerate(time.preps):
            
            # stop on last 
            if i == len(time.preps)-1:
                continue
            
            # look between preps for conjunction
            nxprep = L.d(time.preps[i+1], 'word')
            prepwrds = L.d(prep, 'word')
            iwords = list(range(prepwrds[0], nxprep[0]))
            is_conj = 'conj' in set(F.pdp.v(w) for w in iwords)
            
            is_quant = 'card' in set(F.ls.v(w) for w in iwords)
            
            if (not is_conj or is_quant) and not val_repeat(time):
                return True
            
    
comdir = CXdata(multi, val_comdir, token_advb)

classes['ppm_compound_dir'] = comdir.timeset

prog()
```

### ø Coordinate Time (simple)

```python
def coord_time(time):
    
    if len(time.times) < 2:
        return False
    
    A = time.times[0]
    B = time.times[1]
    inbetween = set(F.pdp.v(w) for w in range(A+1, B))
    
    conds = (
        not time.preps,
        inbetween, 
        inbetween.issubset({'conj', 'art'}),
    )
    if all(conds):
        return True
    
ctime = CXdata(multi, coord_time, token_advb)

classes['øm_coord_time'] = ctime.timeset

prog()
```

### Coordinate Duration Construction

```python
def coord_dur(time):
    
    if not time.quants:
        return False

    # treat cardinal-number durations
    embedded = next((q for q in L.d(time.cx, 'chunk')
                        if F.label.v(q) == 'quant_NP'
                        and F.embed.v(q) == 'true'
                    ), None)
    
    if embedded:
        return True
    
    # treat qual quant duration
    q_quants = [w for w in L.d(time.cx,'word')
                   if F.sem_set.v(w) == 'quant'
                   and F.ls.v(w) != 'card' 
               ]
    
    if (
        len(q_quants) > 1
        and (F.pdp.v(q_quants[1]-1) == 'conj'
                 or F.pdp.v(q_quants[1]-2) == 'conj')
    ):
        return True
    
coord_dur = CXdata(multi, coord_dur, token_advb)

classes['coord_duration'] = coord_dur.timeset

prog()
```

### Singles Preps with Proper Nouns

These are cases that are not truly multi-phrasal time phrases, wherein the additional phrases modify persons who anchor the time phrase. 

```python
def single_nmpr(time):
    
    atoms = L.d(time.cx, 'phrase_atom')
    
    if len(atoms) < 2:
        return False
    
    appo_phrases = atoms[1:]
    appo_words = [w for a in appo_phrases for w in L.d(a, 'word')]
    person_descripts = {'>B/', 'KHN/', 'NBJ>/'}
    
    is_name = lambda w: (
        F.pdp.v(w) == 'nmpr'
        or F.ls.v(w) == 'gntl'
        or F.lex.v(w) in person_descripts
    )

    names = [w for w in appo_words
                 if is_name(w)]
    
    preps = [w for w in appo_words
                if F.pdp.v(w) == 'prep']
    
    if names and not preps:
        return True
    
sg = CXdata(multi, single_nmpr, token_advb)

classes['singles_nmpr'] = sg.timeset

prog()
```
